```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

# Average Traffic Congestion (Hourly)

```python
data = pd.read_csv('../data/brts_traffic_congestion_data.csv')  # Assuming your data is in a CSV file

# Group by hour and calculate the mean traffic congestion percentage
hourly_congestion = data.groupby('time')['traffic_congestion'].mean()

# Plot the traffic congestion trend over hours
plt.figure(figsize=(10, 6))
plt.plot(hourly_congestion.index, hourly_congestion.values, marker='o', linestyle='-')
plt.title('Average Traffic Congestion by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Traffic Congestion (%)')
plt.xticks(range(24))  # Set x-axis ticks to represent each hour
plt.grid(True)
plt.show()
```

# Average Traffic Congestion by Day of the Week

```python
data = pd.read_csv('../data/amts_traffic_congestion_data.csv')  # Assuming your data is in a CSV file

# Group by day of the week and calculate the mean traffic congestion percentage
daily_congestion = data.groupby('day')['traffic_congestion'].mean()

# Plot the traffic congestion trend over days of the week
plt.figure(figsize=(10, 6))
plt.plot(daily_congestion.index, daily_congestion.values, marker='o', linestyle='-')
plt.title('Average Traffic Congestion by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Traffic Congestion (%)')
plt.xticks(range(1, 8), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])  # Set x-axis ticks to represent days of the week
plt.grid(True)
plt.show()
```

## Average Traffic Congestion by Station

```python
data = pd.read_csv('../data/brts_traffic_congestion_data.csv')

# Group by station and calculate the mean traffic congestion percentage for each station
station_congestion = data.groupby('station_name')['traffic_congestion'].mean().sort_values(ascending=False)
# Filter the data for the current day and hour and the specific station
current_day = datetime.datetime.now().weekday() + 1  # Adding 1 to match the 1-indexed day format in the dataset
current_hour = datetime.datetime.now().hour
specific_station = 'Your Station Name'
data_current_time = data[(data['day'] == current_day) & (data['time'] == current_hour) & (data['station_name'] == specific_station)]

# Calculate the average historical congestion for the current day and hour
average_congestion = data_current_time['traffic_congestion'].mean()

# Calculate the current congestion for the specific station
current_congestion = data_current_time['traffic_congestion'].iloc[0]  # Assuming only one row for the specific station

# Calculate the percentage change in congestion
percentage_change = ((current_congestion - average_congestion) / average_congestion) * 100

# Print the results
print(f"Average Congestion for {specific_station} at Current Day and Hour: {average_congestion:.2f}%")
print(f"Current Congestion for {specific_station}: {current_congestion:.2f}%")
print(f"Percentage Change in Congestion for {specific_station}: {percentage_change:.2f}%"='bar')
plt.title('Average Traffic Congestion by Station')
plt.xlabel('Station')
plt.ylabel('Average Traffic Congestion (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('../data/brts_traffic_congestion_data.csv')

# Pivot the data to create a heatmap of traffic congestion by day of the week and hour
pivot_data = data.pivot_table(index='day', columns='time', values='traffic_congestion', aggfunc='mean')

# Plot a heatmap to visualize the average traffic congestion by day of the week and hour
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, cmap='viridis', annot=True, fmt=".1f")
plt.title('Average Traffic Congestion by Day and Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.show()
```
